```python

import numpy as np

from basicts.utils.result import stat_node
from basicts.utils.visualize import plot_line


# 这个文件的目的，是展示测试集上的总和变化
def visualize_sum(file_dir: str, shape):
    """
    展示一个数据集的全局总和
    """
    # (T, N) -> (T)
    data = np.fromfile(file_dir, dtype=np.float32).reshape(shape)[:, :, 0]
    data = np.sum(data, axis=-1)
    count_days = data.shape[0] // 288
    data = data[:count_days * 288]
    data = data.reshape(-1, 288)
    data = np.sum(data, axis=-1)
    plot_line(title="Traffic Flow by day", x=[i for i in range(count_days)],
              ys=[data], y_axis_name="traffic flow", y_names=["daily traffic flow sum"])


file_dir = "/data3/hanzhi/BasicTS/datasets/PEMS08"
shape=(17856, 170, 3)
#visualize_sum(file_dir, shape)
stat_node(file_dir, shape)
```
