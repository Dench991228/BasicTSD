```python
from utils.visualize import plot_line
import numpy as np

x = np.array([i for i in range(150)])
alpha_0 = 5e-4 # 最大学习率
k = 0.3 # 学习率衰减速度
omega = 10 # 达到一半最大学习率的epoch
s = 3 # 平滑率
lr0 = alpha_0/(1+np.exp(-k*(x-omega)))
lr1 = alpha_0/(1+np.exp(-k/s*(x-s*omega)))
old_lr = [1e-3 for i in range(20)] + [1e-4 for i in range(20)] + [1e-5 for i in range(110)]
plot_line(title="Learning Rate Schedule", x_axis_name="epoch", x=x,
          y_axis_name="lr", y_names=['Sigmoid Scheduler', "Old Scheduler"],
          ys=[lr0-lr1+1e-5, old_lr])
```
